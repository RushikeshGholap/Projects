```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

%matplotlib notebook
```

```python
x1 = np.arange(-1,1,0.01)
x2 = np.arange(-1,1,0.01)

W1 = 0.3
W2 = 0.5
B = 0.5

X1, X2 = np.meshgrid(x1,x2)

Y = W1*X1 + W2*X2 + B

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_wireframe(X1,X2,Y)
```

```python
w1 = np.arange(-1,1,0.01)
w2 = np.arange(-1,1,0.01)

W1,W2 = np.meshgrid(w1,w2)

X1 = [0.3,0.2,1]
X2 = [0.5,0.8,0.8]
B = 0.5


fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

for i in range(3):
    Y = W1*X1[i] + W2*X2[i] + B
    ax.plot_wireframe(W1,W2,Y)
```

# 1. Intuition behind Vector Spaces & Weight Spaces.

### 1. Vector space

Each vector is represented in an n dimensional space as a point. The matrix that contains these vectors is of $R^{n \times m}$ where m is the number of samples and n is the vector dimensionality.

This space is the Vector space and the objective of any classification algorithm is to create a decision boundary in the form of a hyperplane in this space (for binary classification).

```python
sigmoid(np.array(list(zip(x1,x2))))
```

```python
%matplotlib notebook

x1 = np.arange(-1,1,0.01)
x2 = np.arange(-1,1,0.01)

X1,X2 = np.meshgrid(x1,x2)

def sigmoid(x):
    return (1/(1+np.exp(-1*x)))

Z = sigmoid(10*X1+11*X2+1)

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(X1,X2,Z)
```

The intuition of a linear combination $w^T.x + b =0$ is that it can be considered as a hyperplane that separates the vector space. $w^T.x + b > 0$ and $w^T.x + b < 0$.

```python
x1 = np.arange(-10,10,0.01)
x2_lin = 0.4*x1+0.2
x2_sig = 1/(1+np.exp(-1*x2_lin))

plt.plot(x1,x2_sig)
```

w1,w2 vector is orthogonal to the hyperplane w1.x1+w2.x2 = 0
