```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(0,1,0.001)
m1 = 4
c1 = 5
m2 = 1
c2 = 0

def sigmoid(x):
    return (1/(1+np.exp(-1*x)))

y1 = sigmoid(m1*x+c1)
y2 = sigmoid(m2*sigmoid(m1*x+c1) + c2)

plt.plot(x,y1)
plt.plot(x,y2)
```
