```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('oversoul.jpg',cv2.IMREAD_COLOR)

b,g,r = cv2.split(img)       # get b,g,r
rgb_img = cv2.merge([r,g,b])     # switch it to rgb
```

```python
cv2.line(rgb_img, (0,0), (400,400), (255,255,255), 15)   #Opencv works with BGR instead of RGB
cv2.rectangle(rgb_img, (200,300), (500,500), (0,255,0),5)
cv2.circle(rgb_img,(250,350), 55, (0,0,255), 10)

plt.imshow(rgb_img)
```

```python
pts = np.array([[100,50],[200,300],[70,200],[500,100]],np.int32)

cv2.polylines(rgb_img,[pts],True,(0,255,255),12)  #True is for connecting final to first points 
plt.imshow(rgb_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(rgb_img, 'OpenCV Editing', (0,130), font, 3, (200,255,255), 5, cv2.LINE_AA)
plt.imshow(rgb_img)
```
