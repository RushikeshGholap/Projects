```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('oversoul.jpg',cv2.IMREAD_GRAYSCALE)  
#IMREAD_GREYSCALE - 0; IMREAD_COLOR - 1; IMREAD_UNCHANGED = -1;


#Showing the image using matplotlib
plt.imshow(img,cmap='gray',interpolation='bicubic')

plt.title('Oversoul')
plt.plot([500,200],[500,600],'c',linewidth=5)
plt.show()
```

```python
cv2.imwrite('oversouls_1.png',img)
```

```python
cap = cv2.VideoCapture(0)

cv2.startWindowThread()
while True:
    ret, frame = cap.read()
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
