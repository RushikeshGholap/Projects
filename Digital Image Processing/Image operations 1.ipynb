```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bgr_img = cv2.imread('oversoul.jpg',cv2.IMREAD_COLOR)

b,g,r = cv2.split(bgr_img)
img = cv2.merge([r,g,b])
plt.imshow(img)
```

```python
px = img[55,55]
px
```

```python
#Region of Image (ROI)

img1 = img.copy()
img1[100:250, 100:250] = [255,255,255]

plt.imshow(img1)
```

```python
img2 = img.copy()

face = img2[350:600,200:550]
img2[0:250,0:350] = face

cv2.rectangle(img2, (0,0), (350,250), (0,0,0),5)

plt.imshow(img2)
```
