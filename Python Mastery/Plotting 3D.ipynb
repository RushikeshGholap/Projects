# Interesting 3D Plots

```python
import numpy as np
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import math

%matplotlib notebook
```

```python
# Symmetric Waves
x = np.arange(-150,150,10)
y = np.arange(-150,150,10)
X,Y = np.meshgrid(x,y)
R = np.sin(5*X)*np.cos(5*Y)/5

Z = R
fig1 = plt.figure()
ax1 = fig1.add_subplot(111,projection='3d')
ax1.plot_surface(X,Y,Z,cmap='coolwarm')
```

```python
# Ripples
x = np.arange(-1,1,0.01)
y = np.arange(-1,1,0.01)
X,Y = np.meshgrid(x,y)
R = np.sin(10*(X**2 + Y**2))/10

Z = R
fig2 = plt.figure()
ax2 = fig2.add_subplot(111,projection='3d')
ax2.plot_surface(X,Y,Z,cmap='coolwarm')
```

```python
# Tunnel
x = np.arange(-150,150,10)
y = np.arange(-150,150,10)
X,Y = np.meshgrid(x,y)
R = (1/(15*(X**2+Y**2)))

Z = R
fig3 = plt.figure()
ax3 = fig3.add_subplot(111,projection='3d')
ax3.plot_wireframe(X,Y,Z)
```

```python
# Tunnel
x = np.arange(-150,150,10)
y = np.arange(-150,150,10)
X,Y = np.meshgrid(x,y)
R = (X**2-Y**2)

Z = R
fig = plt.figure()
ax = fig3.add_subplot(111,projection='3d')
ax.plot_surface(X,Y,Z,cmap='coolwarm')
```
