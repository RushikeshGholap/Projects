```python
from nltk.corpus import gutenberg as gt
gt.fileids()
```

```python
emma = gt.words('austen-emma.txt')
len(emma)
```

```python
for fileid in gt.fileids():
    num_chars = len(gt.raw(fileid))
    num_words = len(gt.words(fileid))
    num_sents = len(gt.sents(fileid))
    num_vocab = len(set([w.lower() for w in gt.words(fileid)]))
    print(int(num_chars/num_words),int(num_words/num_sents),int(num_words/num_vocab),fileid)
```

```python
macbeth_sentences = gt.sents('shakespeare-macbeth.txt')
macbeth_sentences[1037]
```

```python
longest_sentence = max([len(s) for s in macbeth_sentences])
longest_sentence
```

```python
raw = 'hi my name is akshay. I work at ipredictt data science labs. I am a data science enthusiast.'
tokens = nltk.word_tokenize(raw)
tokens
```

```python
sents = nltk.sent_tokenize(raw)
sents
```

```python
text = nltk.Text(tokens)
text.collocations()
text.concordance('data')
```

```python
#NLP pipeline
data = 'blah blah raw text from any data Data source blah'  #Raw data
raw = data[5:-5] #Trim raw data to relevant text
tokens_all = nltk.word_tokenize(raw) #tokenize
tokens = tokens_all[1:] #select tokens of interest
text = nltk.Text(tokens) #convert to nltk.Text for processing
words = [w.lower() for w in text] #normalize words
vocab = sorted(set(words)) #build vocabulary

print(data)
print(raw)
print(tokens_all)
print(tokens)
print(text)
print(words)
print(vocab)
```
