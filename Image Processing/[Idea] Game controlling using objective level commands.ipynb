```python
import numpy as np
import matplotlib.pyplot as plt
import mss
import mss.tools
from bokeh.plotting import figure
from bokeh.io import output_notebook, show, push_notebook
import cv2
import time

output_notebook()
```

```python
with mss.mss() as sct:
    # The screen part to capture
    monitor = {"top": 320, "left": 790, "width": 650, "height": 150}
    
    
    # Grab the data
    frame = np.array(sct.grab(monitor))
    frame=cv2.flip(frame,-1)
    p = figure()
    myImage = p.image_rgba(image=[frame],x=0, y=0, dw=frame.shape[1], dh=frame.shape[0])
    show(p, notebook_handle=True)
    

    while True:
        frame = np.array(sct.grab(monitor))
        frame=cv2.flip(frame,-1)
        myImage.data_source.data['image']=[frame]
        push_notebook()
        time.sleep(0.1)
```
