# Keras-Sklearn Pipeline

```python
import pandas as pd
import missingno as msno
import numpy as np
import os
```

```python
parent = os.path.realpath("..")
df_path = os.path.join(parent, 'Small Datasets', 'auto-mpg.csv')
```

```python
df = pd.read_csv(df_path)
```

```python
df.head()
```

```python
df1 = df[df['horsepower']!='?']
```

```python
df1['horsepower'] = [int(i) for i in df1['horsepower']]
```

```python
df1.describe()
```

```python
y = df1['mpg']
X = df1.drop('mpg', axis=1)
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from keras.layers import Dense, Dropout, Input
from keras.models import Model
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.externals import joblib
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def autompg_model():
    inp = Input((8,))
    x = Dense(16, activation='relu')(inp)
    x = Dropout(0.5)(x)
    x = Dense(16, activation='relu')(x)
    x = Dropout(0.5)(x)
    out = Dense(1)(x)

    model = Model(inp, out)
    model.compile(loss='mse', optimizer='adam')
    return model
```

```python
X.columns
```

```python
cat_cols = ['car name']
num_cols = ['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin']

encoding = ('encoding', OrdinalEncoder(), cat_cols)
standard = ('standard', StandardScaler(), num_cols)

steps = [standard, encoding]
ct = ColumnTransformer(steps, remainder='passthrough')

preprocess = ('preprocess', ct)
regress = ('reg', KerasRegressor(build_fn=autompg_model, epochs=500, verbose=1, validation_split=0.15))
steps1 = [preprocess, regress]

pipe = Pipeline(steps1)
```

```python
pipe
```

```python
pipe.fit(X, y)
```

```python
test = X[X.index==100]
test
```

```python
df1[df1.index==100]
```

```python
pipe.predict(test)
```

```python
model_path = os.path.join(parent, 'Small Datasets', 'model.pkl')
model_path
```

```python
joblib.dump(pipe, model_path)
```

```python
mod = joblib.load(model_path)
```

```python
mean_absolute_error(y, mod.predict(X))
```

```python
np.sqrt(mean_squared_error(y, mod.predict(X)))
```

```python
r2_score(y, mod.predict(X))
```
