```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

from sklearn.neighbors import KNeighborsClassifier
%matplotlib inline
```

```python
iris_dataset = load_iris()
print(iris_dataset.keys())
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_dataset.data,iris_dataset.target, random_state=0)

#train_test_split creates a list of 4 ndarrays which can be stored respectively as above in 4 ndarrys separately
type(X_train)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
iris_dataframe = pd.DataFrame(X_train, columns = iris_dataset.feature_names)
g = pd.scatter_matrix(iris_dataframe, c=y_train, figsize = (8,8), alpha = 0.8, diagonal = 'kde')
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train, y_train)
```

```python
#Test prediction - 
randomtest = np.array([[5,4,1,2]])
prediction = knn.predict(randomtest)
print(format(iris_dataset.target_names[prediction]))
```

```python
#Evaluating model
y_pred = knn.predict(X_test)
print(format(y_pred))
print(format(np.mean(y_pred == y_test)))
```

```python
#One step for accuracy evaluation
print('Accuracy of model: ',format(knn.score(X_test, y_test)))
```
