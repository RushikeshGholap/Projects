```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 100
```

```python
x = np.random.random((n,))
y = np.random.random((n,))
dx = np.random.random((n,))
dy = np.random.random((n,))
```

```python
theta = np.arctan(dy/dx)
```

```python
ndx = np.sin(theta)/10
```

```python
ndy = np.cos(theta)/10
```

```python
plt.figure()
plt.xlim(0,1.25)
plt.ylim(0,1.25)
for i in range(len(x)):
    plt.arrow(x[i],y[i],ndx[i],ndy[i],head_width=0.02, length_includes_head=True)
```

```python
def draw_field(f1,f2,o1,o2,raw=False):
    #each of the input parameters needs to be a 1d array, of same length, scaled 0-1
    dx = o1-f1
    dy = o2-f2
    
    theta = np.arctan(dy/dx)
    ndx = np.multiply(np.sign(dx),np.cos(theta))/10
    ndy = np.multiply(np.sign(dy),np.sin(theta))/10
    
    plt.figure(figsize=(10,10))
    plt.xlim(-2.5,2.5)
    plt.ylim(-2.5,2.5)
    
    if raw==True:
        for i in range(len(x)):
            plt.arrow(f1[i],f2[i],dx[i],dy[i],head_width=0.03, length_includes_head=True)
    else:
        for i in range(len(x)):
            plt.arrow(f1[i],f2[i],ndx[i],ndy[i],head_width=0.03, length_includes_head=True)
```

```python
from sklearn.datasets import make_circles
```

```python
F,O = make_circles(n_samples=n)
```

```python
f1, f2, o1, o2 = F[:,0], F[:,1], O, np.zeros((n,))
```

```python
draw_field(f1,f2,o1,o2)
```
